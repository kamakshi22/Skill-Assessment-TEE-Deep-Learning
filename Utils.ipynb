```python
import glob
import os,sys
from PIL import Image
from numpy import *
import Imageio

path = '/Users/kamakshibansal/UCL/PROJECT/original_data/*.avi'   
files= glob.glob(path) 
outpath = '/Users/kamakshibansal/UCL/PROJECT/original_data_frames/'
box1 = (234, 107, 412, 215)
box2 = (739, 108, 980, 215)
box3 = (853, 904, 1200,1000)
for file in files: 
    
#     print (file)
    vid = imageio.get_reader(file, 'ffmpeg')
#     fps = vid.get_meta_data()['fps']
    for i, im in enumerate(vid):
        img = Image.fromarray(im)
        
        #img = imageio.imwrite(file +'%i.jpg' %(i),im[:, :, :])
        #print(img.getdata())
        #print (im.shape)
        img.paste(0, box1)
        img.paste(0, box2)
        img.paste(0, box3)
        #plt.imshow(img)
        #plt.show()
        #print(file)
        filename = file[file.find('data/',)+5:file.find('.avi')]
        img.save(outpath+ filename+'_%i.jpg' %(i))
        
```

```python
import numpy as np

X=[0,20,40,60,80,100]
Y=[0,100/6, 200/6, 50, 400/6, 500/6, 100]
Z=[0,25,50,75,100]


Score_5_cri= []
Score_6_cri= []
Score_4_cri= []

for x in X:
    for j in X:
        for i in X:
            k=(i+j+x)/3
            Score_5_cri.append(k)
            
            
for x in Y:
    for j in Y:
        for i in Y:
            h=(i+j+x)/3
            Score_6_cri.append(h)  

            
for x in Z:
    for j in Z:
        for i in Z:
            k=(i+j+x)/3
            Score_4_cri.append(k)            

Score_4_cri.sort()
Score_4= np.unique(Score_4_cri)

Score_5_cri.sort()
Score_5= np.unique(Score_5_cri)

Score_6_cri.sort()
Score_6=np.unique(Score_6_cri)


def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]

def Score_processing(clas, value):
    
    if (clas==1 or clas==2 or clas==3 or clas==6 or clas==8 or clas==9):
        correct_val= find_nearest(Score_5, value)
    elif (clas==5 or clas==10):
        correct_val= find_nearest(Score_6, value)
    else:
        correct_val= find_nearest(Score_4, value)
        
    return correct_val 


file=pd.read_csv("/Users/kamakshibansal/UCL/PROJECT/Predicted_from_Alex.csv")
correct_val=[]
for clas, Criteria_Pct in file[['Class','Criteria_Pct']].values:
    correct= Score_processing(clas, Criteria_Pct)
    correct_val.append(correct)
    
file["Rectified_value"]= correct_val 
file.to_csv("Corrected_Predicted_from_Alex.csv", encoding='utf-8')

            
            

```
